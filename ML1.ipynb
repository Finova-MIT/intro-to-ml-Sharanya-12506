```python
import numpy as np

matrix1 = np.random.randint(1, 11, size=(3, 3))
matrix2 = np.random.randint(1, 11, size=(3, 3))

print("Matrix 1:\n", matrix1)
print("\nMatrix 2:\n", matrix2)

matrix_addition = np.add(matrix1 , matrix2)
print("\nMatrix Addition (Matrix 1 + Matrix 2):\n", matrix_addition)

matrix_subtraction = np.subtract(matrix1 , matrix2)
print("\nMatrix Subtraction (Matrix 1 - Matrix 2):\n", matrix_subtraction)

elt_wise_mult = np.matmul(matrix1 , matrix2)
print("\nElement-wise Matrix Multiplication :\n", elt_wise_mult)

dot_product = np.dot(matrix1, matrix2)
print("\nDot Product (Matrix 1 . Matrix 2):\n", dot_product)

trans_matrix1 = np.transpose(matrix1)
trans_matrix2 = np.transpose(matrix2)
print("\nTranspose of Matrix 1:\n", trans_matrix1)
print("\nTranspose of Matrix 2:\n", trans_matrix2)

det_matrix1 = np.linalg.det(matrix1)
det_matrix2 = np.linalg.det(matrix2)

print("\nDeterminant of Matrix 1:", det_matrix1)
print("Determinant of Matrix 2:", det_matrix2)

# Inverse can be found if determinant is non-zero
if det_matrix1 != 0:
    inv_matrix1 = np.linalg.inv(matrix1)
    print("\nInverse of Matrix 1:\n", inv_matrix1)

else:
    print("\nMatrix 1 is not invertible.")

if det_matrix2 != 0:
    inv_matrix2 = np.linalg.inv(matrix2)
    print("\nInverse of Matrix 2:\n", inv_matrix2)
else:
    print("\nMatrix 2 is not invertible.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = "Cereals.csv"  # Ensure the file is in the same directory
df = pd.read_csv(file_path)

# Data Cleaning: Replace '?' with NaN and convert columns to numeric
df.replace("?", pd.NA, inplace=True)
df["carbo"] = pd.to_numeric(df["carbo"])
df["sugars"] = pd.to_numeric(df["sugars"])
df["potass"] = pd.to_numeric(df["potass"])

# Fill missing values with median
df.fillna(df.select_dtypes(include=["number"]).median(), inplace=True)

# Unique cereal brands
unique_brands = df["name"].nunique()
print(f"Unique cereal brands: {unique_brands}")
```

```python
# Cereals per manufacturer
manufacturer_counts = df["mfr"].value_counts()
print("\nCereals per manufacturer:\n", manufacturer_counts)

# Visualization: Cereals per manufacturer
plt.figure(figsize=(8, 5))
sns.barplot(x=manufacturer_counts.index, y=manufacturer_counts.values, palette="viridis")
plt.xlabel("Manufacturer")
plt.ylabel("Number of Cereals")
plt.title("Number of Cereals per Manufacturer")
plt.show()
```

```python
# Hot vs Cold Cereals
type_counts = df["type"].value_counts()
print("\nHot vs Cold Cereals:\n", type_counts)

# Visualization: Hot vs Cold Cereals
plt.figure(figsize=(5, 5))
sns.barplot(x=type_counts.index, y=type_counts.values, palette="coolwarm")
plt.xlabel("Cereal Type (C = Cold, H = Hot)")
plt.ylabel("Count")
plt.title("Hot vs Cold Cereals")
plt.show()
```

```python
# Best and Worst Cereal
best_cereal = df.loc[df["rating"].idxmax(), ["name", "rating"]]
worst_cereal = df.loc[df["rating"].idxmin(), ["name", "rating"]]
print("\nBest Cereal:\n", best_cereal)
print("\nWorst Cereal:\n", worst_cereal)
```

```python
# Average rating for Hot vs Cold Cereals
rating_comparison = df.groupby("type")["rating"].mean()
print("\nAverage Rating: Hot vs Cold Cereals:\n", rating_comparison)

# Visualization: Rating Comparison
plt.figure(figsize=(5, 5))
sns.barplot(x=rating_comparison.index, y=rating_comparison.values, palette="coolwarm")
plt.xlabel("Cereal Type")
plt.ylabel("Average Rating")
plt.title("Average Rating: Hot vs Cold Cereals")
plt.show()
```

```python
# Highest Fiber and Lowest Sugar Cereal
highest_fiber = df.loc[df["fiber"].idxmax(), ["name", "fiber"]]
lowest_sugar = df.loc[df["sugars"].idxmin(), ["name", "sugars"]]
print("\nCereal with Highest Fiber:\n", highest_fiber)
print("\nCereal with Lowest Sugar:\n", lowest_sugar)
```

# **12. Combining plots**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("matches.csv")

df.fillna(method='ffill', inplace=True)

df['date'] = pd.to_datetime(df['date'])

matches_per_date = df.groupby('date').size()

fig, axes = plt.subplots(3, 1, figsize=(12, 18))

# Line plot - Time series of matches
axes[0].plot(matches_per_date.index, matches_per_date.values, marker='o', linestyle='-')
axes[0].set_xlabel('Date')
axes[0].set_ylabel('Number of Matches')
axes[0].set_title('Number of Cricket Matches Over Time')
axes[0].tick_params(axis='x', rotation=45)
axes[0].grid(True)

# Bar plot - Matches per season
matches_per_season = df.groupby('season').size()
axes[1].bar(matches_per_season.index, matches_per_season.values, color='skyblue')
axes[1].set_xlabel('Season')
axes[1].set_ylabel('Number of Matches')
axes[1].set_title('Number of Matches Per Season')
axes[1].tick_params(axis='x', rotation=45)
axes[1].grid(axis='y')

# Scatter plot - Example: Matches per venue (assuming venue exists in dataset)
venue_counts = df['venue'].value_counts()
axes[2].scatter(venue_counts.index, venue_counts.values, color='red')
axes[2].set_xlabel('Venue')
axes[2].set_ylabel('Number of Matches')
axes[2].set_title('Number of Matches Per Venue')
axes[2].tick_params(axis='x', rotation=90)
axes[2].grid(True)

plt.tight_layout()
plt.show()

# Generating a correlation heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("titanic.csv")

print("Dataset Info:")
df.info()

print("\nFirst 5 Rows:")
print(df.head())

df.fillna(method='ffill', inplace=True)

print("\nSummary Statistics:")
print(df.describe())

# Countplot of Survived column
plt.figure(figsize=(6, 4))
sns.countplot(x='Survived', data=df, palette='coolwarm')
plt.title("Survival Count")
plt.show()

# Distribution of Age
plt.figure(figsize=(8, 5))
sns.histplot(df['Age'].dropna(), bins=30, kde=True, color='blue')
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()

# Survival rate
plt.figure(figsize=(8, 5))
sns.barplot(x='Pclass', y='Survived', data=df, ci=None, palette='viridis')
plt.title("Survival Rate by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Survival Rate")
plt.show()

# Correlation heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("titanic.csv")

print("Dataset Info:")
df.info()

print("\nFirst 5 Rows:")
print(df.head())

df.fillna(df.median(numeric_only=True), inplace=True)
for col in df.select_dtypes(include=['object']).columns:
    df[col].fillna(df[col].mode()[0], inplace=True)

print("\nSummary Statistics:")
print(df.describe())

plt.figure(figsize=(6, 4))
sns.countplot(x='Survived', data=df, palette='coolwarm')
plt.title("Survival Count")
plt.show()

plt.figure(figsize=(8, 5))
sns.histplot(df['Age'], bins=30, kde=True, color='blue')
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()

plt.figure(figsize=(8, 5))
sns.barplot(x='Pclass', y='Survived', data=df, ci=None, palette='viridis')
plt.title("Survival Rate by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Survival Rate")
plt.show()

plt.figure(figsize=(8, 5))
sns.barplot(x='Sex', y='Survived', data=df, ci=None, palette='pastel')
plt.title("Survival Rate by Gender")
plt.xlabel("Gender")
plt.ylabel("Survival Rate")
plt.show()

plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()

sns.pairplot(df[['Survived', 'Pclass', 'Age', 'Fare']].dropna(), hue='Survived', palette='coolwarm')
plt.show()
```
